```python
def count_words(sentence):
    words=sentence.lower().split()
    word_count={}
    for word in words:
        word = word.strip(",.!?")
        if word in word_count:
            word_count[word]+=1
        else:
            word_count[word]=1
    return word_count
sentence="Hello we we you us "
word_freq=count_words(sentence)
print(word_freq)
```

```python
class Node:
    def _init_(self, data):
        self.data = data
        self.next = None

# Define the LinkedList class
class LinkedList:
    def _init_(self):
        self.head = None

    # Function to append a new node
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    # Function to print the linked list
    def print_list(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    # Function to reverse the linked list
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next  # Save next
            current.next = prev       # Reverse current node's pointer
            prev = current            # Move pointers one position ahead
            current = next_node
        self.head = prev

# Example usage
llist = LinkedList()
llist.append(1)
llist.append(2)
llist.append(3)
llist.append(4)

print("Original Linked List:")
llist.print_list()

llist.reverse()

print("Reversed Linked List:")
llist.print_list()
```

```python
#binary search
def binary_search(arr,target):
    low,high=0,len(arr)-1
    while low<=high:
        mid=(low+high)
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            low=mid+1
        else:
            high=mid-1
    return -1
data=[10,20,30,40,50]
print(binary_search(data,50))
        
        
```

```python
def bubble_sort(arr):
    for n in range(len(arr)-1,0,-1):
        swapped=False
        for i in range(n):
            if arr[i]>arr[i+1]:
                arr[i],arr[i+1]=arr[i+1],arr[i]
                swapped=True
        if not swapped:
            break
arr=[6,7,2,1,4,5,0]
print("Unsorted list is:")
print(arr)
bubble_sort(arr)
print("sorted list is:")
print(arr)
```

```python
def max_coins(chests):
    max_sum = chests[0]
    current_sum = chests[0]
    
    for coins in chests[1:]:
        # Either start a new subarray at this chest or extend the current one
        current_sum = max(coins, current_sum + coins)
        max_sum = max(max_sum, current_sum)
    
    return max_sum

# Example usage
coins_list = [2, -3, 4, -1, 2, 1, -5, 4]
print(max_coins(coins_list))  # Output: 6
```

```python
def special_sum(arr):
    # Edge cases
    if not arr or len(arr) <= 3:
        return 0

    # Extract even and odd positioned elements
    even_pos = [arr[i] for i in range(0, len(arr), 2)]
    odd_pos = [arr[i] for i in range(1, len(arr), 2)]

    # Sort them
    even_pos.sort()
    odd_pos.sort()

    # Find required elements
    second_largest_even = even_pos[-2] if len(even_pos) >= 2 else 0
    second_smallest_odd = odd_pos[1] if len(odd_pos) >= 2 else 0

    return second_largest_even + second_smallest_odd


# Example
arr = [3, 2, 1, 7, 5, 4]
print(special_sum(arr))  # Output: 7
```

```python
def length_of_longest_substring(s):
    char_set=set()
    left=0
    max_length=0
    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left+=1
            char_set.add(s[right])
        max_length=max(max_length,right-left+1)
    return max_length
print(length_of_longest_substring("abcabcbb"))
```

```python
def minimize_coin(amount,coins):
    coins.sort(reverse=True)
    change=[]
    for coin in coins:
        while coin<=amount:
            amount-=coin
            change.append(coin)
    return change
amount=120
coins=[2,5,10,20,50]
purse=minimize_coin(amount,coins)
print("coins used:",purse)
        
```
