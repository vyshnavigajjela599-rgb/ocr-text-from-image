```python
#prime number
n = int(input("Enter a number: "))
if n <= 1:
    print(n, "is not prime")
else:
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print(n, "is not prime")
            break
    else:
        print(n, "is prime")
```

```python
s="aba"
n=len(s)
ans=""
for i in range(n-1,-1,-1):
    ans=ans+s[i]
if ans==s:
    print("Palindrome")
else:
    print("No")
```
